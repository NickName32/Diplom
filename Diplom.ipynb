# Вычисление углов

```python
def calculateangle(a, b, c):
    a = np.array(a)  # First
    b = np.array(b)  # Mid
    c = np.array(c)  # End

    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians*180.0/np.pi)

    if angle > 180.0:
        angle = 360-angle
        
    return angle
```

```python
cap = cv2.VideoCapture(0)

# Curl counter variables
counter = 0
stage = None


# Setup mediapipe instance
with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5, model_complexity=1) as pose:
    while cap.isOpened():
        ret, frame = cap.read()
        
        # Recolor image to RGB
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False

        # Make detection
        results = pose.process(image)

        # Recolor back to BGR
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)


        # Extract landmarks
        try:
            landmarks = results.pose_landmarks.landmark

            # Get coordinates
            l_shoulder = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x, landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
            l_elbow = [landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].x, landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].y]
            l_wrist = [landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].x, landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].y]
            
            # Calculate angle
            left_elbow_angle = calculateangle(l_shoulder, l_elbow, l_wrist)
            
            # Visualize angle
            cv2.putText(image, str(left_elbow_angle), 
                           tuple(np.multiply(l_elbow, [640, 480]).astype(int)),
                           cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA
                                )
            
            # Curl counter logic
            if left_elbow_angle > 160:
                stage = "down"
            if left_elbow_angle < 30 and stage == 'down':
                stage="up"
                counter+=1
                print(counter) 

        except:
            pass

        # Render curl counter
        # Setup status box
        cv2.rectangle(image, (0,0), (250,73), (245,117,16), -1)

        # Rep data
        cv2.putText(image, 'REPS', (15,12), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,0), 1, cv2.LINE_AA)
        cv2.putText(image, str(counter), 
                    (10,60), 
                    cv2.FONT_HERSHEY_SIMPLEX, 2, (255,255,255), 2, cv2.LINE_AA)

        # Stage data
        cv2.putText(image, 'STAGE', (90,12), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,0), 1, cv2.LINE_AA)
        cv2.putText(image, stage, 
                    (90,60), 
                    cv2.FONT_HERSHEY_SIMPLEX, 2, (255,255,255), 2, cv2.LINE_AA)

        
        # Render detections
        mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                                 mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=2),
                                 mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2)
                                 )

        cv2.imshow('Mediapipe Feed', image)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import mediapipe as mp

mp_pose = mp.solutions.pose
mp_drawing = mp.solutions.drawing_utils

def calculate_angle(a, b, c):
    a = np.array(a)
    b = np.array(b)
    c = np.array(c)
    
    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians*180.0/np.pi)
    
    if angle > 180.0:
        angle = 360 - angle
        
    return angle

cap = cv2.VideoCapture(0)

# Инициализация переменных
counter = 0
stage = None
view = "side"
active_arm = None
prev_stage = None
frame_count = 0
view_history = []

# Настройки
MIN_DIP_ANGLE = 80  # Минимальный угол сгиба для нижней позиции
MAX_DIP_ANGLE = 160  # Максимальный угол разгиба для верхней позиции
SHOULDER_ANGLE_THRESHOLD = 30  # Порог угла плеча

with mp_pose.Pose(min_detection_confidence=0.8, min_tracking_confidence=0.8) as pose:
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            continue
            
        frame_count += 1
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False
        results = pose.process(image)
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        try:
            landmarks = results.pose_landmarks.landmark
            
            # Получаем координаты ключевых точек
            nose = [landmarks[mp_pose.PoseLandmark.NOSE.value].x,
                   landmarks[mp_pose.PoseLandmark.NOSE.value].y]
            
            l_shoulder = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x,
                         landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
            r_shoulder = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x,
                         landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]
            
            l_elbow = [landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].x,
                      landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].y]
            r_elbow = [landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].x,
                      landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].y]
            
            l_wrist = [landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].x,
                      landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].y]
            r_wrist = [landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].x,
                      landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].y]
            
            l_hip = [landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x,
                    landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y]
            r_hip = [landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x,
                    landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y]
            
            # Вычисляем углы
            left_elbow_angle = calculate_angle(l_shoulder, l_elbow, l_wrist)
            right_elbow_angle = calculate_angle(r_shoulder, r_elbow, r_wrist)
            
            left_shoulder_angle = calculate_angle(l_elbow, l_shoulder, l_hip)
            right_shoulder_angle = calculate_angle(r_elbow, r_shoulder, r_hip)
            
            # Определяем ракурс по положению носа относительно плеч
            shoulder_center_x = (l_shoulder[0] + r_shoulder[0]) / 2
            if abs(nose[0] - shoulder_center_x) < 0.1:
                current_view = "front"
            else:
                current_view = "side"
            
            # Фильтруем определение ракурса
            view_history.append(current_view)
            if len(view_history) > 10:
                view_history.pop(0)
            
            # Определяем ракурс по большинству в истории
            if view_history.count("front") > view_history.count("side"):
                view = "front"
            else:
                view = "side"
            
            # Определяем активную руку
            if view == "front":
                # Для фронтального вида выбираем руку с более согнутым локтем
                if left_elbow_angle < right_elbow_angle:
                    active_arm = "left"
                    elbow_angle = left_elbow_angle
                    shoulder_angle = left_shoulder_angle
                else:
                    active_arm = "right"
                    elbow_angle = right_elbow_angle
                    shoulder_angle = right_shoulder_angle
            else:
                # Для вида сбоку выбираем руку, которая ближе к камере
                if l_elbow[0] > r_elbow[0]:
                    active_arm = "left"
                    elbow_angle = left_elbow_angle
                    shoulder_angle = left_shoulder_angle
                else:
                    active_arm = "right"
                    elbow_angle = right_elbow_angle
                    shoulder_angle = right_shoulder_angle
            
            # Логика подсчета повторений
            if elbow_angle < MIN_DIP_ANGLE and shoulder_angle > SHOULDER_ANGLE_THRESHOLD:
                current_stage = "down"
            elif elbow_angle > MAX_DIP_ANGLE and shoulder_angle < SHOULDER_ANGLE_THRESHOLD:
                current_stage = "up"
            else:
                current_stage = stage
            
            # Счетчик повторений
            if current_stage == "up" and stage == "down":
                counter += 1
                print(f"Rep: {counter} ({view} view, {active_arm} arm)")
            
            stage = current_stage
            
            # Отображаем углы
            cv2.putText(image, f"L: {int(left_elbow_angle)}", 
                       tuple(np.multiply(l_elbow, [image.shape[1], image.shape[0]]).astype(int)),
                       cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA)
            
            cv2.putText(image, f"R: {int(right_elbow_angle)}", 
                       tuple(np.multiply(r_elbow, [image.shape[1], image.shape[0]]).astype(int)),
                       cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA)
            
        except Exception as e:
            print(f"Error: {e}")
            pass

        # Отображаем информацию
        info_text = f"REPS: {counter} | STAGE: {stage if stage else 'None'}"
        arm_text = f"ARM: {active_arm if active_arm else 'None'}"
        view_text = f"VIEW: {view.upper()}"
        
        cv2.rectangle(image, (0, 0), (image.shape[1], 100), (245, 117, 16), -1)
        cv2.putText(image, info_text, (10, 30), 
                   cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 0), 2, cv2.LINE_AA)
        cv2.putText(image, arm_text, (10, 60), 
                   cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 0), 2, cv2.LINE_AA)
        cv2.putText(image, view_text, (10, 90), 
                   cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 0), 2, cv2.LINE_AA)

        # Рисуем landmarks
        if results.pose_landmarks:
            mp_drawing.draw_landmarks(
                image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                mp_drawing.DrawingSpec(color=(245, 117, 66), thickness=2, circle_radius=2),
                mp_drawing.DrawingSpec(color=(245, 66, 230), thickness=2, circle_radius=2)
            )

        cv2.imshow('Accurate Dip Counter', image)
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import mediapipe as mp

# Инициализация MediaPipe Pose
mp_pose = mp.solutions.pose
mp_drawing = mp.solutions.drawing_utils

def calculate_angle(a, b, c):
    """Вычисляет угол между тремя точками"""
    a = np.array(a)
    b = np.array(b)
    c = np.array(c)
    
    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians*180.0/np.pi)
    
    if angle > 180.0:
        angle = 360 - angle
        
    return angle

cap = cv2.VideoCapture(0)

# Counter variables
counter = 0
stage = None
active_arm = None
valid_position = False

with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    while cap.isOpened():
        ret, frame = cap.read()
        
        if not ret:
            continue
            
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False
        results = pose.process(image)
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        try:
            landmarks = results.pose_landmarks.landmark

            # Получаем координаты для обеих рук
            l_shoulder = landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value]
            l_elbow = landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value]
            l_wrist = landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value]
            
            r_shoulder = landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value]
            r_elbow = landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value]
            r_wrist = landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value]
            
            nose = landmarks[mp_pose.PoseLandmark.NOSE.value]
            l_hip = landmarks[mp_pose.PoseLandmark.LEFT_HIP.value]
            r_hip = landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value]

            # Вычисляем углы для обеих рук
            left_elbow_angle = calculate_angle([l_shoulder.x, l_shoulder.y], 
                                              [l_elbow.x, l_elbow.y], 
                                              [l_wrist.x, l_wrist.y])
            
            right_elbow_angle = calculate_angle([r_shoulder.x, r_shoulder.y], 
                                               [r_elbow.x, r_elbow.y], 
                                               [r_wrist.x, r_wrist.y])
            
            left_shoulder_angle = calculate_angle([l_elbow.x, l_elbow.y],
                                                [l_shoulder.x, l_shoulder.y],
                                                [l_hip.x, l_hip.y])
            
            right_shoulder_angle = calculate_angle([r_elbow.x, r_elbow.y],
                                                 [r_shoulder.x, r_shoulder.y],
                                                 [r_hip.x, r_hip.y])

            # Определяем активную руку по тому, какая рука ниже (ближе к выполнению отжимания)
            active_arm = 'left' if l_elbow.y > r_elbow.y else 'right'
            
            if active_arm == 'left':
                elbow_angle = left_elbow_angle
                shoulder_angle = left_shoulder_angle
                elbow_y = l_elbow.y
                elbow_x = l_elbow.x
                elbow_point = (int(l_elbow.x * image.shape[1]), int(l_elbow.y * image.shape[0]))
            else:
                elbow_angle = right_elbow_angle
                shoulder_angle = right_shoulder_angle
                elbow_y = r_elbow.y
                elbow_x = r_elbow.x
                elbow_point = (int(r_elbow.x * image.shape[1]), int(r_elbow.y * image.shape[0]))

            # Проверка правильного положения
            nose_above_elbow = nose.y < elbow_y  # Нос выше локтя
            elbow_proper_x = (active_arm == 'left' and elbow_x < nose.x) or (active_arm == 'right' and elbow_x > nose.x)
            valid_position = nose_above_elbow and elbow_proper_x

            # Логика подсчета повторений
            if valid_position:
                if elbow_angle > 160 and shoulder_angle < 20:
                    stage = "up"
                elif elbow_angle < 90 and shoulder_angle > 30 and stage == "up":
                    stage = "down"
                    counter += 1
                    print(f"Reps: {counter} ({active_arm} arm)")

        except Exception as e:
            print(f"Error: {e}")
            pass

        # Отрисовка landmarks и информации
        if results.pose_landmarks:
            mp_drawing.draw_landmarks(
                image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                mp_drawing.DrawingSpec(color=(245,117,66), 
                mp_drawing.DrawingSpec(color=(245,66,230))
            )

        # Отображение углов
        cv2.putText(image, f"L: {int(left_elbow_angle)}", 
                   (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255,255,255), 2)
        cv2.putText(image, f"R: {int(right_elbow_angle)}", 
                   (image.shape[1]-100, 50), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255,255,255), 2)

        # Информационная панель
        cv2.rectangle(image, (0,0), (300, 120), (245,117,16), -1)
        
        cv2.putText(image, 'REPS', (15,25), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,0,0), 2)
        cv2.putText(image, str(counter), (15,80), cv2.FONT_HERSHEY_SIMPLEX, 1.5, (255,255,255), 2)
        
        cv2.putText(image, 'STAGE', (100,25), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,0,0), 2)
        cv2.putText(image, stage if stage else "None", (100,80), cv2.FONT_HERSHEY_SIMPLEX, 1.2, (255,255,255), 2)
        
        cv2.putText(image, 'ARM', (200,25), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,0,0), 2)
        cv2.putText(image, active_arm if active_arm else "None", (200,80), 
                   cv2.FONT_HERSHEY_SIMPLEX, 1.2, (255,255,255), 2)

        # Индикатор положения
        pos_color = (0,255,0) if valid_position else (0,0,255)
        pos_text = "GOOD" if valid_position else "BAD POS"
        cv2.circle(image, elbow_point, 10, pos_color, -1)
        cv2.putText(image, pos_text, (elbow_point[0]+15, elbow_point[1]), 
                   cv2.FONT_HERSHEY_SIMPLEX, 0.5, pos_color, 2)

        cv2.imshow('Dip Counter', image)
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from cvzone.PoseModule import PoseDetector

def calculate_angle(a, b, c):
    # Функция для вычисления угла между тремя точками
    a, b, c = np.array(a), np.array(b), np.array(c)
    radians = np.arctan2(c[1] - b[1], c[0] - b[0]) - np.arctan2(a[1] - b[1], a[0] - b[0])
    angle = np.abs(radians * 180.0 / np.pi)
    return angle if angle <= 180 else 360 - angle

cap = cv2.VideoCapture(0)
detector = PoseDetector(detectionCon=0.8, trackCon=0.8)

counter = 0
stage = "up"
angle_buffer = []
buffer_size = 4

# Пороговые значения углов для отжиманий
min_angle = 60  # угол для верхнего положения
max_angle = 170  # угол для нижнего положения

while True:
    success, img = cap.read()
    if not success:
        break

    img = cv2.resize(img, (1000, 600))
    img = detector.findPose(img, draw=False)  # Отключаем отрисовку основного скелета
    lmList, _ = detector.findPosition(img, draw=False)  # Получаем список точек

    if lmList:
        try:
            # Получаем координаты плеч, локтей и запястий
            left_shoulder = lmList[11][1:3]
            left_elbow = lmList[13][1:3]
            left_wrist = lmList[15][1:3]

            right_shoulder = lmList[12][1:3]
            right_elbow = lmList[14][1:3]
            right_wrist = lmList[16][1:3]

            # Вычисляем углы для локтей
            left_angle = calculate_angle(left_shoulder, left_elbow, left_wrist)
            right_angle = calculate_angle(right_shoulder, right_elbow, right_wrist)

            # Средний угол
            avg_angle = (left_angle + right_angle) / 2

            # Логика подсчета повторений
            if avg_angle > max_angle:  # Рука прямая
                stage = "up"
            elif avg_angle < min_angle and stage == "up":  # Рука сгибается
                stage = "down"
                counter += 1  # Увеличиваем счетчик

            # Отрисовка линий (плечо - локоть - запястье)
            cv2.line(img, tuple(map(int, left_shoulder)), tuple(map(int, left_elbow)), (255, 0, 0), 4)
            cv2.line(img, tuple(map(int, left_elbow)), tuple(map(int, left_wrist)), (255, 0, 0), 4)
            cv2.line(img, tuple(map(int, right_shoulder)), tuple(map(int, right_elbow)), (0, 0, 255), 4)
            cv2.line(img, tuple(map(int, right_elbow)), tuple(map(int, right_wrist)), (0, 0, 255), 4)

            # Отображение углов
            cv2.putText(img, f"L: {int(left_angle)}", tuple(map(int, left_elbow)), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)
            cv2.putText(img, f"R: {int(right_angle)}", tuple(map(int, right_elbow)), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)

        except Exception as e:
            print(f"Ошибка: {e}")

    # Отображение счетчика отжиманий
    cv2.rectangle(img, (20, 20), (250, 100), (0, 0, 0), -1)
    cv2.putText(img, f'Pushups: {counter}', (30, 70), cv2.FONT_HERSHEY_SIMPLEX, 1.5, (0, 255, 0), 3)
    cv2.putText(img, f'Stage: {stage}', (30, 120), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 0), 2)

    # Показ изображения
    cv2.imshow("Push-Up Counter", img)

    # Выход по клавише 'q'
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
